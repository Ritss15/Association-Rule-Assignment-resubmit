```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_excel("Online retail.xlsx", header=None)

df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
transactions = []

for i in range(len(df)):
    transactions.append(df.iloc[i,0].split(','))
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

basket = pd.DataFrame(te_array, columns=te.columns_)
```

```python
frequent_items = apriori(
    basket,
    min_support=0.01,
    use_colnames=True
)

frequent_items.head()
```

```python
rules = association_rules(
    frequent_items,
    metric="confidence",
    min_threshold=0.3
)

rules[['antecedents','consequents',
       'support','confidence','lift']]
```

```python
sns.scatterplot(
    x="support",
    y="confidence",
    size="lift",
    data=rules
)

plt.show()
```

Analysis (Write Below Output)The association rules reveal relationships between products.
High lift values indicate strong customer purchasing patterns.
These insights help businesses improve recommendation systems and marketing strategies.
